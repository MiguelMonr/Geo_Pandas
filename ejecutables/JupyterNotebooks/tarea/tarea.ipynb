Paso 7. Agrupa las escuelas por su alcaldia. Una vez hecho eso crea una nueva geometria que "envuelva" a las escuelas de cada alcaldia y graficalos

El paso anterior guardalo como una capa y graficalo en conjunto a la grafica del paso 1

```python

# Supongamos que tienes un GeoDataFrame llamado gdf con una columna 'alcaldi'

# Agrupar por la columna 'alcaldi'
gdf_grupo = gdf.groupby('alcaldi')

# Lista para almacenar las geometrías de envolventes
geom = []

# Crear envolventes convexas para cada grupo
for alcaldia, grupo in gdf_grupo:
    g_alcaldia = MultiPoint(grupo['geometry'].tolist()).convex_hull
    geom.append({'alcaldia': alcaldia, 'geometry': g_alcaldia})

# Crear una nueva GeoDataFrame a partir de las geometrías
gdf_envolv = gpd.GeoDataFrame(geom, geometry='geometry')

# 4. Plotear únicamente las envolventes
ax = gdf.plot(color='lightgray', figsize=(10, 10))  # Plotear el fondo con geometrías originales
gdf_envolv.plot(ax=ax, edgecolor='red', facecolor='none', linewidth=2)

# Configurar el gráfico
ax.set_title('Envolturas Convexas por Alcaldía')
ax.set_xlabel('Longitud')
ax.set_ylabel('Latitud')
ax.legend(["Geometrías Originales", "Envolturas Convexas"], loc='upper right')

# Mostrar el gráfico
plt.show()
```
