```python
unique_values = df['alcaldi'].unique()
geo_data = []
ax = df.plot(color='blue', label='Main data')
for value in unique_values:
    filtered_df = df[df['alcaldi'] == value]
    new_geometry = filtered_df['geometry'].unary_union
    gpd.GeoDataFrame(geometry=[new_geometry]).plot(ax=ax)
```
