Paso 5. Crea una nueva columna en el gdf del paso 1 que se llame distancia; después escoge una escuela del data frame y que la nueva columna tenga la distancia de todas las escuelas a la que escogiste 

```python

escuela_referencia = df_escuelas.geometry.iloc[0]

df_escuelas['distancia'] = df_escuelas.geometry.distance(escuela_referencia)

df_escuelas
```

Paso 7. Agrupa las escuelas por su alcaldia. Una vez hecho eso crea una nueva geometria que "envuelva" a las escuelas de cada alcaldia y graficalos

```python
import pandas as pd
import matplotlib.pyplot as plt


grupos = df_escuelas.groupby('alcaldi')

hulls = gpd.GeoDataFrame(columns=['alcaldi', 'geometry'])

for name, group in grupos:
    hull = group.geometry.unary_union.convex_hull
    hulls = pd.concat([hulls, gpd.GeoDataFrame([[name, hull]], columns=['alcaldi', 'geometry'])], ignore_index=True)

hulls.set_crs(df_escuelas.crs, inplace=True)

hulls.plot()
plt.show()
```

El paso anterior guardalo como una capa y graficalo en conjunto a la grafica del paso 1

```python
hulls.to_file('convex_hulls.geojson', driver='GeoJSON')
```

```python

hulls = gpd.read_file('convex_hulls.geojson')

ax = df_escuelas.plot(color='blue', markersize=5, label='Escuelas')

hulls.plot(ax=ax, color='none', edgecolor='red', label='Convex Hulls')

plt.legend()

plt.show()
```
