Paso 1. Guarda el geojson en un dataframe e imprimelo. 

```python
import geopandas as gpd

url = 'escuelas_privadas.geojson'
df = gpd.read_file(url)
df
```

Paso 2. Crea otro gdf con solo las escuelas en la delegacion Miguel Hidalgo

```python
# Filtra las escuelas en la delegación Miguel Hidalgo
delegacion_miguel_hidalgo = df[df['alcaldi'] == 'MIGUEL HIDALGO']
delegacion_miguel_hidalgo
```

Paso 3. Setea el indice para que ahora sea la colonia

```python
# Paso 3: Setea el índice para que ahora sea la colonia
delegacion_miguel_hidalgo = delegacion_miguel_hidalgo.set_index('colonia')

# Imprime el nuevo GeoDataFrame con el índice modificado
delegacion_miguel_hidalgo
```

Paso 3. Grafica el dataframe

```python
# Paso 3: Grafica el DataFrame
delegacion_miguel_hidalgo.plot()

```

Paso 4. Crea dos nuevas columnas en el gdf original que sean para el area y el perimetro, en ellas haz tu calculo de cada una.

```python
# Paso 4: Crear dos nuevas columnas para área y perímetro en el gdf original
target_crs = 'EPSG:3395'
df = df.to_crs(target_crs)  
df['area'] = df.geometry.area
df['perimetro'] = df.geometry.length
df
```

Paso 5. Crea una nueva columna en el gdf del paso 1 que se llame distancia; después escoge una escuela del data frame y que la nueva columna tenga la distancia de todas las escuelas a la que escogiste 

```python
df['distancia'] = 0.0  # Inicializa la columna con valores predeterminados

# Selecciona la escuela de referencia
escuela_referencia = df[(df['nombre'] == 'CENDI PART SEDI LOMAS')].iloc[0]

# Punto de referencia (geometry) de la escuela seleccionada
punto_referencia = escuela_referencia['geometry']

# Calcula la distancia entre la escuela de referencia y todas las demás escuelas
df['distancia'] = df['geometry'].distance(punto_referencia)

df
```

Paso 6. Realiza una consulta para saber cuales son las 3 escuelas mas lejos del punto que elegiste, debe imprimir solamente el nombre de las escuelas y su distancia al punto

```python
# Ordena el GeoDataFrame por distancia en orden descendente
df_sorted = df.sort_values(by='distancia', ascending=False)

# Selecciona las tres escuelas más lejanas (distintas)
escuelas_lejanas = df_sorted.drop_duplicates(subset='nombre').iloc[:3]

# Imprime el nombre de las escuelas y su distancia al punto de referencia
for idx, row in escuelas_lejanas.iterrows():
    print(f"Escuela: {row['nombre']}, Distancia: {row['distancia']} km")
```

Paso 7. Agrupa las escuelas por su alcaldia. Una vez hecho eso crea una nueva geometria que "envuelva" a las escuelas de cada alcaldia y graficalos

```python

# Agrupa las escuelas por su alcaldía
grouped_by_alcaldia = df.groupby('alcaldi')['geometry'].agg(lambda x: gpd.GeoSeries(x).unary_union)

# Crea un nuevo GeoDataFrame con las geometrías agrupadas
gdf_alcaldias = gpd.GeoDataFrame(geometry=grouped_by_alcaldia)

# Grafica los polígonos que envuelven a las escuelas de cada alcaldía utilizando Geopandas
gdf_alcaldias.plot(edgecolor='black', linewidth=1, alpha=0.5, figsize=(10, 8), legend=True, cmap='Set3')
gdf_alcaldias


```

El paso anterior guardalo como una capa y graficalo en conjunto a la grafica del paso 1

```python
#supongo que debo graficar el df inicial del paso 1
# Guarda el GeoDataFrame resultante como un archivo GeoJSON
gdf_alcaldias.to_file('capa.geojson', driver='GeoJSON')
df.plot()
gdf_alcaldias.plot(edgecolor='black', linewidth=1, alpha=0.5, figsize=(10, 8), legend=True, cmap='Set3')
```
