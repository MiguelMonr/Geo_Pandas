Geopandas permite la lecuta y analisis de archivos geograficos, tales como lo pueden ser los shapefiles, archivos de forma y geojson 

La lectura la puede hacer desde un url 

```python
url = "http://d2ad6b4ur7yvpq.cloudfront.net/naturalearth-3.3.0/ne_110m_land.geojson"
df = gpd.read_file(url)
df
```

Al ser geopandas una extensión de pandas la manipulación de datos es similar a la que se hace en los dataframes o series. 

```python
df.drop(columns=['featureclass'], inplace=True)
df
```

A diferencia de otros metodos en python, para leer archivos en geopandas debemos de usar el path del archivo. Para eso utilizamos la biblioteca os

```python
import os
```

Obteniendo el path. 

```python

# Ruta actual
ruta_actual = os.getcwd()  # Obtiene la ruta actual donde estás trabajando

# Ruta al archivo buscado en otra carpeta
ruta_archivo_buscado = os.path.join(ruta_actual, '794551067260_s', 'conjunto_de_datos' , 'a1206_ii.shp')

gdf = gpd.read_file(ruta_archivo_buscado)
gdf
```

También permite la lectura de archivos comprimidos 

```python
ruta= "794551067260_s.zip"

ruta_absoluta_zip = os.path.abspath(ruta)

df= gpd.read_file('zip://' + ruta_absoluta_zip + '!conjunto_de_datos/a1206_ii.shp') 
df
```

Una manera más rapida de obtener el path sin usar la biblioteca geodatasets

```python
import geodatasets as gds

#GeoData de la ciudad de Nueva York
#Long path 

gdf = gpd.read_file(
    gds.get_path("geoda.nyc"))

gdf
```

Los shapefiles pueden llegar muchos datos, miles, por lo que al leer un shp puedes decidir que datos son los que quieras tener

```python
gdf = gpd.read_file(
    gds.get_path("geoda.nyc")
    ,rows= 10)

gdf
```

Puedes decidir incluso que campos deseas 

```python

gdf = gpd.read_file(
    gds.get_path("geoda.nyc"),
    include_fields=["name", "rent2008", "kids2000"],
)

gdf
```

```python
gdf = gpd.read_file(
    gds.get_path("geoda.nyc"),
    where="subborough='Coney Island'",
)

gdf
```

```python
from shapely.geometry import Point
# Crear un conjunto de datos de puntos
data = {
    'Nombre': ['Punto 1', 'Punto 2', 'Punto 3'],
    'ID': [1, 2, 3],
    'geometry': [Point(-74.006, 40.7128), Point(-73.986, 40.7484), Point(-73.977, 40.7815)]
}

# Crear un GeoDataFrame a partir de los datos de puntos
gdf = gpd.GeoDataFrame(data, crs='EPSG:4326')

# Escribir el GeoDataFrame a un archivo GeoJSON
gdf.to_file("puntos.geojson", driver='GeoJSON')

# Mostrar la información del GeoDataFrame
print(gdf)
```
